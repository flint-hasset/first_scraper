```python
# import libraries
#!pip install BeautifulSoup4
#!pip install urllib3
import urllib
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
```

```python
http = urllib3.PoolManager()
```

```python
url_2 = 'https://countryeconomy.com/ratings/canada'
response = http.request('GET', url_2)
soup_countryeconomy = BeautifulSoup(response.data, 'lxml')
```

```python
url = 'https://tradingeconomics.com/canada/rating'
response = http.request('GET', url)
soup = BeautifulSoup(response.data, 'lxml')
```

```python
table = soup.find('table',{'class':'table table-hover'})
```

```python
dummy_array = np.zeros((40,4))
new_table = pd.DataFrame(dummy_array)
row_marker = 0
for row in table.find_all('tr'):
    column_marker = 0
    columns = row.find_all('td')
    for column in columns:
        new_table.iloc[row_marker,column_marker] = column.get_text()
        column_marker += 1
    row_marker += 1
new_table
```
